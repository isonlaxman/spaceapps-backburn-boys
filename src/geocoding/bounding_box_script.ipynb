```python
import pandas as pd
import geopy
from geopy.geocoders import Nominatim
```

```python
def space_out(y):
    if y == "PearceRAAF":
        return "Pearce RAAF"
    else:
        return ''.join(map(lambda x: x if x.islower() else " "+x, y))
```

```python
df = pd.read_csv("../data/weatherAUS.csv")
df = df.drop_duplicates(subset=["Location"])
df["Spaced Location"] = df["Location"].apply(space_out)
df
```

```python
locator = Nominatim(user_agent="spaceapps", timeout=10)
```

```python
bounding_boxes = {}

for i, r in df.iterrows():
    try:
        bounding_boxes[r["Location"]] = locator.geocode(r["Spaced Location"] + ", Australia").raw["boundingbox"]
        print(r["Location"])
    except:
        print("FAIL: " + r["Location"])

bounding_boxes
```
